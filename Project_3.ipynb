```python
# Top five Netflix country distribution separated by type of release(i.e Movies,TV shows), visualizing via Bar graph
country_type_counts=df.groupby(['country','type']).size().reset_index(name='count')
top_countries = country_type_counts.groupby('country')['count'].sum().nlargest(5).index
filtered_data = country_type_counts[country_type_counts['country'].isin(top_countries)]
plt.figure(figsize=(8,4))
sns.barplot(data=filtered_data, x='country', y='count', hue='type', palette='pastel')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 5 Countries on Netflix by type of release')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```

```python
# Top 10 genres of movies and TV Shows on Netflix
df['genre'] = df['listed_in'].str.split(', ')
df = df.explode('genre')
movies_df = df[df['type'] == 'Movie']
tv_shows_df = df[df['type'] == 'TV Show']
top_10_movie_genres = movies_df['genre'].value_counts().head(10)
top_10_tv_genres = tv_shows_df['genre'].value_counts().head(10)
fig, axes = plt.subplots(2,1, figsize=(8, 8))
top_10_movie_genres.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Top 10 Movies Genre on Netflix')
axes[0].set_xlabel('Genre')
axes[0].set_ylabel('Count')
axes[0].set_xticklabels(top_10_movie_genres.index, rotation=90)
top_10_tv_genres.plot(kind='bar', ax=axes[1], color='orange')
axes[1].set_title('Top 10 TV Shows Genre on Netflix')
axes[1].set_xlabel('Genre')
axes[1].set_ylabel('Count')
axes[1].set_xticklabels(top_10_tv_genres.index, rotation=90)

plt.tight_layout()
plt.show()



```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['Month'] = df['date_added'].dt.month
df['Day of Week'] = df['date_added'].dt.day_name()
release_counts = df.groupby(['Month', 'Day of Week']).size().unstack()
print("Aggregated release countsby month and day of the week:")
print(release_counts)
plt.figure(figsize=(8,6))
sns.heatmap(release_counts, annot=True, fmt='d', cmap='YlGnBu', cbar=True, linewidth=0.6)
plt.title('Aggregated Release Counts by Month and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Month')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
